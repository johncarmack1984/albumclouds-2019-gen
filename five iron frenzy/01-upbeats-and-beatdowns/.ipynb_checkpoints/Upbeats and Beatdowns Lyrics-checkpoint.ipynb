```python
import codecs
```

```python
f = codecs.open("upbeats-and-beatdowns-lyrics.txt", 'r')
```

```python
text = f.read()
```

```python
import numpy as np
from PIL import Image
```

```python
album_coloring = np.array(Image.open('upbeats-mask-2560.png'))
```

```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
image_colors = ImageColorGenerator(album_coloring)
wordcloud = WordCloud(
    background_color="#00adef",
    width=2560,
    height=2560,
    mask=album_coloring,
    font_path='ShadowsIntoLight.ttf').generate(text).recolor(color_func=image_colors)
wordcloud.to_file('upbeats-wordcloud.png')

plt.figure(figsize=(16,16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.figure()
```
