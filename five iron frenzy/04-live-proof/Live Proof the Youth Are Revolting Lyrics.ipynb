```python
import codecs
```

```python
import numpy as np
from PIL import Image
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
import matplotlib.image as mpimg
```

```python
f = codecs.open("live-proof-lyrics.txt", 'r')
text = f.read()
```

```python
clean_text = text
#.upper()
#.replace('CENSORED','')
#.lower().replace('god','God').replace('jesus','Jesus')
```

```python
album_coloring = np.array(Image.open('live-proof-cover.png'))
image_colors = ImageColorGenerator(album_coloring)
wordcloud = WordCloud(
    background_color=None, #"#5e4951",
    mode="RGBA",
    width=2560,
    height=2560,
    mask=album_coloring,
    font_path='FrederickatheGreat-Regular.ttf',
    collocations=False,
    stopwords=STOPWORDS).generate(clean_text).recolor(color_func=image_colors)
wordcloud.to_file('live-proof-wordcloud.tiff')

#img=mpimg.imread('hype-cover-border.png')
#plt.imshow(img)
plt.figure(figsize=(16,16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
