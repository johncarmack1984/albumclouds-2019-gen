```python
import codecs
```

```python
import numpy as np
from PIL import Image
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

**Disc 1**

```python
f = codecs.open("./end-is-here-disc-1-lyrics.txt", 'r')
text_one = f.read()
```

```python
clean_text_one = text_one.replace('uh','')
#.lower().replace('god','God').replace(
#    'combat chuck','Combat Chuck').replace('america','America')
```

```python
album_coloring = np.array(Image.open('end-is-here-disc-one.png'))
image_colors = ImageColorGenerator(album_coloring)
wordcloud = WordCloud(
    background_color=None, #"#00adef",
    mode="RGBA",
    width=2560,
    height=2560,
    mask=album_coloring,
    font_path='./Montserrat-SemiBold.ttf',
    collocations=False,
    normalize_plurals=False,
    prefer_horizontal=0.99,
    stopwords=STOPWORDS).generate(clean_text_one).recolor(color_func=image_colors)
wordcloud.to_file('end-is-here-disc-one.tiff')

plt.figure(figsize=(16,16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.figure()
```

```python
from textblob import TextBlob
analysis_one = TextBlob(clean_text_one)
print ('{} words'.format(len(analysis_one.words)))
print ('')
print ('polarity: {}'.format(analysis_one.sentiment.polarity))
print ('subjectivity: {}'.format(analysis_one.sentiment.subjectivity))
```
