```python
import codecs
```

```python
f = codecs.open("hype-lyrics.txt", 'r')
```

```python
text = f.read()
```

```python
import numpy as np
from PIL import Image
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
import matplotlib.image as mpimg
```

```python
clean_text = text
#.upper()
#.replace('CENSORED','')
#.lower().replace('god','God').replace('jesus','Jesus')
```

```python
album_coloring = np.array(Image.open('hype-cover.png'))
image_colors = ImageColorGenerator(album_coloring)
wordcloud = WordCloud(
    background_color="#f7920e",
    width=2560,
    height=2560,
    mask=album_coloring,
    font_path='Lora-Regular.ttf',
    collocations=False,
    stopwords=STOPWORDS).generate(clean_text).recolor(color_func=image_colors)
wordcloud.to_file('hype-wordcloud.tiff')

#img=mpimg.imread('hype-cover-border.png')
#plt.imshow(img)
plt.figure(figsize=(16,16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.figure()
```
