```python
import codecs
```

```python
f = codecs.open("quantity-lyrics.txt", 'r')
```

```python
text = f.read()
```

```python
import numpy as np
from PIL import Image
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
clean_text = text.replace('CENSORED','')
#.lower().replace('god','God').replace('jesus','Jesus')
```

```python
album_coloring = np.array(Image.open('quantity-cover.png'))
image_colors = ImageColorGenerator(album_coloring)
wordcloud = WordCloud(
    background_color=None, # "#682b1b",
    mode="RGBA",
    width=2560,
    height=2560,
    mask=album_coloring,
    font_path='RalewayDots-Regular.ttf',
    collocations=False,
    stopwords=STOPWORDS).generate(clean_text).recolor(color_func=image_colors)
wordcloud.to_file('quantity-wordcloud.tiff')

plt.figure(figsize=(16,16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.figure()
```

```python
from textblob import TextBlob
analysis = TextBlob(clean_text)
print ('{} words'.format(len(analysis.words)))
print ('')
print ('polarity: {}'.format(analysis.sentiment.polarity))
print ('subjectivity: {}'.format(analysis.sentiment.subjectivity))
```
