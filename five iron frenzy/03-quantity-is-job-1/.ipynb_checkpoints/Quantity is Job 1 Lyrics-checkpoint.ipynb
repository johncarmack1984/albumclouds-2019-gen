```python
import codecs
```

```python
f = codecs.open("newest-album-ever-lyrics.txt", 'r')
```

```python
text = f.read()
```

```python
import numpy as np
from PIL import Image
```

```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
clean_text = text.lower().replace('god','God').replace('jesus','Jesus')
```

```python
album_coloring = np.array(Image.open('quantity-cover.png'))
image_colors = ImageColorGenerator(album_coloring)
wordcloud = WordCloud(
    background_color="#801c22",
    width=2560,
    height=2560,
    mask=album_coloring,
    font_path='Risque-Regular.ttf').generate(clean_text).recolor(color_func=image_colors)
wordcloud.to_file('quantity-wordcloud.png')

plt.figure(figsize=(16,16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.figure()
```
