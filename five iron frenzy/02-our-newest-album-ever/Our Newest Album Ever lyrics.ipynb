```python
import codecs
```

```python
f = codecs.open("newest-album-ever-lyrics.txt", 'r')
```

```python
text = f.read()
```

```python
import numpy as np
from PIL import Image
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
clean_text = text
#.lower().replace('micah','Micah').replace('canada','Canada').replace(
#    'god','God').replace('jesus','Jesus').replace(
#    'holden caufield','Holden Caufield').replace(
#    'jack kerouac','Jack Kerouac')
```

```python
album_coloring = np.array(Image.open('newest-cover.png'))
image_colors = ImageColorGenerator(album_coloring)
wordcloud = WordCloud(
    background_color= None, #"#801c22",
    mode="RGBA",
    width=2560,
    height=2560,
    mask=album_coloring,
    font_path='./Risque-Regular.ttf',
    collocations=False,
    normalize_plurals=False,
    prefer_horizontal=0.98,
    stopwords=STOPWORDS).generate(clean_text).recolor(color_func=image_colors)
wordcloud.to_file('newest-ever-wordcloud.tiff')

plt.figure(figsize=(16,16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.figure()
```
