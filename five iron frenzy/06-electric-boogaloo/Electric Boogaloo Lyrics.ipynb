```python
import codecs
```

```python
import numpy as np
from PIL import Image
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
import matplotlib.image as mpimg
```

```python
f = codecs.open("boogaloo-lyrics.txt", 'r')
text = f.read()
```

```python
clean_text = text
#.upper()
#.replace('CENSORED','')
#.lower().replace('god','God').replace('jesus','Jesus')
```

```python
album_coloring = np.array(Image.open('boogaloo-cover.png'))
image_colors = ImageColorGenerator(album_coloring)
wordcloud = WordCloud(
    background_color= None, #"#3c3d39", #"#595b55", #'white', #
    mode="RGBA", 
    width=2560,
    height=2560,
    mask=album_coloring,
    font_path='Lato-Light.ttf',
    collocations=False,
    prefer_horizontal=0.999,
    stopwords=STOPWORDS,
    normalize_plurals=False).generate(clean_text).recolor(color_func=image_colors)
wordcloud.to_file('boogaloo-wordcloud.tiff')

#img=mpimg.imread('hype-cover-border.png')
#plt.imshow(img)
plt.figure(figsize=(16,16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.figure()
```
