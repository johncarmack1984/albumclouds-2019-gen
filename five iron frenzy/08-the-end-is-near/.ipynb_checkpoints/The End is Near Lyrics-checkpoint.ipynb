```python
import codecs
```

```python
import numpy as np
from PIL import Image
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
f = codecs.open("./end-is-near-lyrics.txt", 'r')
```

```python
text = f.read()
```

```python
clean_text = text
#.lower().replace('god','God').replace(
#    'combat chuck','Combat Chuck').replace('america','America')
```

```python
album_coloring = np.array(Image.open('end-is-near-cover.png'))
image_colors = ImageColorGenerator(album_coloring)
wordcloud = WordCloud(
    background_color=None, #"#00adef",
    mode="RGBA",
    width=2560,
    height=2560,
    mask=album_coloring,
    font_path='./Montserrat-SemiBold.ttf',
    collocations=False,
    normalize_plurals=False,
    prefer_horizontal=0.94,
    stopwords=STOPWORDS).generate(clean_text).recolor(color_func=image_colors)
wordcloud.to_file('end-is-near-wordcloud.tiff')

plt.figure(figsize=(16,16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.figure()
```
